#### Importing Required Libraries 

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
### Loading Data using Pandas

BMS_dataset = pd.read_csv("D:/Machine Learning Projects Files/Big Mart Prediction/train_data.csv")
BMS_dataset.head()
```

```python
### Splitting the X and Y Values

X = BMS_dataset.drop("Item_Outlet_Sales", axis = 1)
y = BMS_dataset.Item_Outlet_Sales
```

```python
### Now Creating new 3 Binary columns using BaseEstimator

from sklearn.base import BaseEstimator

class outletTypeEncoder(BaseEstimator):
    
    def __init__(self):
        pass
    
    def fit(self, documents, y = None):
        return self
    
    def transform(self, x_dataset):
        x_dataset['Outlet_grocery_store'] = (x_dataset["Outlet_Type"] == "Grocery Store") * 1
        x_dataset['outlet_supermarket_3'] = (x_dataset["Outlet_Type"] == 'Supermarket Type3') * 1
        x_dataset['outlet_identifier_OUT027'] = (x_dataset["Outlet_Identifier"] == 'OUT027') * 1
        
        return x_dataset
```

```python
pre_process = ColumnTransformer(remainder = 'passthrough', 
                               transformers = [('drop_Columns', 'drop', ['Item_Identifier', 'Item_Fat_Content', 
                                                            'Item_Type', 'Outlet_Identifier', 
                                                            'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']),
                                             ('Weight_imputer', SimpleImputer(strategy = 'mean'), ["Item_Weight"]), 
                                             ('Scale', StandardScaler(), ['Item_MRP'])])
```

```python
### Creating Pipeline 

from sklearn.pipeline import Pipeline
model_Pipeline = Pipeline(steps = [('getoutlet_binary', outletTypeEncoder()),
                                  ('preprocessing', pre_process),
                                  ('model', RandomForestRegressor())
                                  ])

model_Pipeline.fit(X, y)

model_Pipeline.predict(X)
```
