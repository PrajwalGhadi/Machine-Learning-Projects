## Digit Data Analysis By Support Vector Machine(SVM)

### Importing Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

### Loading Dataset

```python
data_set = load_digits()
```

```python
data_set.target
```

```python
dir(data_set)
```

### Creating DataFrame 

```python
df = pd.DataFrame(data_set.data, data_set.target)
df.head()
```

```python
### Adding Target to dataset
```

```python
df['target'] = data_set.target
df.head(20)
```

### Creating X and Y for Training and Testing of Data

```python
X = df.drop('target', axis = 'columns')
y = df.target
```

### Creating Model using train_test_split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
y_train.head()
```

### Creating Model of SVM

```python
SVM_model = SVC(kernel = 'rbf')
```

```python
SVM_model.fit(X_train, y_train)
```

```python
SVM_model.score(X_test, y_test)
```

```python
Predicted = SVM_model.predict(X_test)
```

### Creating Final Dataset of Actual and Predicted Values

```python
Final_Data = pd.DataFrame({'Actual Value' : y_test, 'Predicted Va'})
```
